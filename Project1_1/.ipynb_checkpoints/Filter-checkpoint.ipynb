```python
import sys

# any(title.startswith(prefix) for prefix in prefixes)

blacklist_suffix = (
".png", "gif", ".jpg", ".jpeg", ".tiff", ".tif", ".xcf", ".mid", ".ogg", ".ogv", ".svg", ".djvu", ".oga", ".flac",
".opus", ".wav", ".webm", ".ico", ".txt", ".ico")

blacklist_special = ("404.php", "Main_Page", "-")

namespace = ['Media', 'Special', 'Talk', 'User', 'User_talk', 'Wikipedia', 'Wikipedia_talk', 'File', 'File_talk',
             'MediaWiki', 'MediaWiki_talk', 'Template', 'Template_talk', 'Help', 'Help_talk', 'Category',
             'Category_talk', 'Portal', 'Portal_talk', 'Book', 'Book_talk', 'Draft', 'Draft_talk', 'Education_Program',
             'Education_Program_talk', 'TimedText', 'TimedText_talk', 'Module', 'Module_talk', 'Gadget', 'Gadget_talk',
             'Gadget_definition', 'Gadget_definition_talk']

blacklist_namespace = []

for name in namespace:
    name_lower = name.lower()
    blacklist_namespace.append(name_lower)

def decode(encoded):
    def getHexValue(b):
        if '0' <= b <= '9':
            return chr(ord(b) - 0x30)
        elif 'A' <= b <= 'F':
            return chr(ord(b) - 0x37)
        elif 'a' <= b <= 'f':
            return chr(ord(b) - 0x57)
        return None

    if encoded is None:
        return None

    encodedChars = encoded
    encodedLength = len(encodedChars)
    decodedChars = ''
    encodedIdx = 0

    while encodedIdx < encodedLength:
        if encodedChars[encodedIdx] == '%' and encodedIdx + 2 < encodedLength and getHexValue(encodedChars[encodedIdx + 1]) and getHexValue(encodedChars[encodedIdx + 2]):
            #  current character is % char
            value1 = getHexValue(encodedChars[encodedIdx + 1])
            value2 = getHexValue(encodedChars[encodedIdx + 2])
            decodedChars += chr((ord(value1) << 4) + ord(value2))
            encodedIdx += 2
        else:
            decodedChars += encodedChars[encodedIdx]
        encodedIdx += 1
    return str(decodedChars)
#p

fname="pageviews-20161109-000000"
print('mark1')
# open file

#use the following line will give encoding bug

#filein = open(fname, "rt", encoding='utf-8')

# Read and return a list of lines from the stream.
rows = sys.stdin.readlines()
print('mark2')
valid_count=0;
all_count=0;

dic=dict() # dictionary for collect the data ("title", count)

for r in rows:

    all_count+=1

    if not (r.startswith('en') or r.startswith('en.m')):
        continue

    print(r.rstrip())

    # r =r.encode("utf-8")
    # r=decode(r)
    row = r.split()

    if len(row) != 4:
        continue


    domain_code = row[0];
    page_title = row[1];
    count_views = row[2];
    total_response_size = row[3];
    
    
    if not (domain_code == 'en' or domain_code == 'en.m'):
        continue

    # def any(iterable):
    #     for element in iterable:
    #         if element:
    #             return True
    #     return False

    if any(page_title.lower().startswith(name) for name in blacklist_namespace):
        continue

    if page_title[0].islower():
        continue

    if any(page_title.lower().endswith(suffix) for suffix in blacklist_suffix):
        continue

    if page_title.lower().endswith("_(disambiguation)"):
        continue

    if any(page_title == special for special in blacklist_special):
        continue


    valid_count+=1

    #merge en and en.m
    if (domain_code=='en'):
        domain_code='en.m'

     #put the valid page_title as key and view count as value to dictionary
    if page_title not in dic:
        dic[page_title]=int(count_views)
    else:
        #dic[page_title].append(int(count_views))
        #adding value existing value
        dic[page_title]=dic[page_title]+int(count_views)
    print (dic)


print("The number of all rows are", all_count, '\n')
print("The number of all valid rows are", valid_count , '\n')

# This will sort the dictionary by the values of each entry within the dictionary from smallest to largest.
# it returns an ordered list of (key, value) tuples.
# https://stackoverflow.com/questions/613183/how-do-i-sort-a-dictionary-by-value
final=sorted(dic.items(),key=lambda x:x[1],reverse=True)

output = open('output', 'wt')

for (key,value) in final:
    output.write(key+"\t"+str(value)+ '\n')

output.close()
```
